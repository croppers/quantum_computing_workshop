```python
import numpy as np
import qiskit

def visualize(quantum_circuit):
    # run the quantum circuit on the simulator
    sim = qiskit.Aer.get_backend('aer_simulator')
    result = sim.run(quantum_circuit).result()
    counts = result.get_counts()

    # visualize the results
    return qiskit.visualization.plot_histogram(counts) # plot them using a histogram
```

# Initializing a quantum state
### classical
constructing the circuits

```python
qc_0 = qiskit.QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state_0 = [1,0] # initial state is |0>
qc_0.initialize(initial_state_0, 0) # apply |0> to the first qubit
qc_0.save_statevector()
qc_0.draw()  # draw the circuit

qobj_0 = qiskit.assemble(qc_0)

qc_1 = qiskit.QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state_1 = [0,1]   # initial state is |1>
qc_1.initialize(initial_state_1, 0) # apply |1> to the second qubit
qc_1.save_statevector()
qc_1.draw()  # draw the circuit

qobj_1 = qiskit.assemble(qc_1)
```

#### making the measurements
circuit 0

```python
sim = qiskit.Aer.get_backend('aer_simulator')  # use the classical Aer simulator
result_0 = sim.run(qobj_0).result()
output_0 = result_0.get_statevector()
print(output_0)

result_0 = sim.run(qobj_0).result()
counts_0 = result_0.get_counts()
qiskit.visualization.plot_histogram(counts_0)
```

circuit 1

```python
# circuit 1
sim = qiskit.Aer.get_backend('aer_simulator')  # use the classical Aer simulator
result_1 = sim.run(qobj_0).result()
output_1 = result_0.get_statevector()
print(output_1)

result_1 = sim.run(qobj_1).result()
counts_1 = result_1.get_counts()
qiskit.visualization.plot_histogram(counts_1)
```

### quantum
real values

```python
qc_3 = qiskit.QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state_3 = [1/np.sqrt(2),1/np.sqrt(2)] # initial state is |q_3>
qc_3.initialize(initial_state_3, 0) # apply |q_3> to the first qubit
qc_3.save_statevector()
qc_3.draw()  # draw the circuit

qobj_3 = qiskit.assemble(qc_3)

sim = qiskit.Aer.get_backend('aer_simulator')  # use the classical Aer simulator
result_3 = sim.run(qobj_3).result()
output_3 = result_3.get_statevector()
print(output_3)

result_3 = sim.run(qobj_3).result()
counts_3 = result_3.get_counts()
qiskit.visualization.plot_histogram(counts_3)
```

imaginary values

```python
qc_3 = qiskit.QuantumCircuit(1)  # Create a quantum circuit with one qubit
initial_state_3 = [1j/np.sqrt(2),1j/np.sqrt(2)] # initial state is |q_3>
qc_3.initialize(initial_state_3, 0) # apply |q_3> to the first qubit
qc_3.save_statevector()
qc_3.draw()  # draw the circuit

qobj_3 = qiskit.assemble(qc_3)

sim = qiskit.Aer.get_backend('aer_simulator')  # use the classical Aer simulator
result_3 = sim.run(qobj_3).result()
output_3 = result_3.get_statevector()
print(output_3)

result_3 = sim.run(qobj_3).result()
counts_3 = result_3.get_counts()
qiskit.visualization.plot_histogram(counts_3)
```
